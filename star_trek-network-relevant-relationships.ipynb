```python
%%time
total = 0
count = 0
error = 0
content = ''
with open('wikidata-20150907-star_trek.json') as f:
    content += '[\n'
    for line in f:
        try:
            total += 1
            cleaned = line.strip()[:-1]
            d = json.loads(cleaned) # keys: ['type', 'labels', 'descriptions', 'claims', 'aliases', 'sitelinks', 'id']
            for rel in list(d['claims']):
                if rel not in only_relevant_rels_dict:
                    del(d['claims'][rel])
            if d['claims']:
                if 'en' in d['descriptions']:
                    val = d['descriptions']['en']
                    d['descriptions'].clear()
                    d['descriptions']['en'] = val
                else:
                    d['descriptions'].clear()
                if 'en' in d['labels']:
                    val = d['labels']['en']
                    d['labels'].clear()
                    d['labels']['en'] = val
                else:
                    d['labels'].clear()
                if 'en' in d['aliases']:
                    val = d['aliases']['en']
                    d['aliases'].clear()
                    d['aliases']['en'] = val
                else:
                    d['aliases'].clear()
                if 'enwiki' in d['sitelinks']:
                    val = d['sitelinks']['enwiki']
                    d['sitelinks'].clear()
                    d['sitelinks']['enwiki'] = val
                else:
                    d['sitelinks'].clear()
                content += json.dumps(d) + ',\n'
                count += 1
        except Exception as e:
            error += 1
    content += ']'
with open('wikidata-20150907-star_trek-relevant-rels.json', 'w') as f:
    f.write(content)
print('Total:', total)
print('Count:', count)
print('Error:', error)
```
